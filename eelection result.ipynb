```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2024 election result.csv")
parties = df['Party']
won = df['Won']
```

```python
plt.figure(figsize=(6, 6))
plt.pie(won, labels=None, autopct='%1.1f%%', startangle=140)
plt.title('Election Results by Party')
plt.legend(parties, loc='upper center', bbox_to_anchor=(0.5, .05), ncol=2, title='Party')
plt.show()
```
